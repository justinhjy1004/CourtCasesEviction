```python
import re
def order_continued(actions):

    if actions is None:
        return False
    
    y = re.findall("order-continuance", actions, flags=re.IGNORECASE)

    if len(y) > 0:
        return True
    else:
        return False
```

```python
def motion_continued(actions):

    if actions is None:
        return False
    
    y = re.findall("motion-continuance", actions, flags=re.IGNORECASE)

    if len(y) > 0:
        return True
    else:
        return False
```

```python
df.ORDER = df.ACTIONS.apply(order_continued)
```

```python
df.MOTION = df.ACTIONS.apply(motion_continued)
```

```python
order_cont = df[df.ORDER == True]
```

```python
order_cont.sample(10).index
```
